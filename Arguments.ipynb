Function arguments

```python
#default argument

def myfun(id ,name = "adhi"):
  print("ID" ,id)
  print("Name" ,name)
  return
myfun(15)
```

```python
def add_numbers(a = 7 ,b = 8):
  sum = a + b
  print('Sum:' ,sum)
add_numbers(2)
```

```python
#keyword arguments

def display_info(firstname ,lastname):
  print('firstname:' ,firstname)
  print('lastname:' ,lastname)
display_info(lastname = 'cartman' ,firstname ='john')
```

```python
#variable-length arguments

def find_sum(*numbers):
  result = 0

  for num in numbers:
    result = result + num

  print("sum =" ,result)

find_sum(1,2,3,7,8,9)
```

```python
def find_product(*numbers):
  result = 1

  for num in numbers:
    result = result * num

  print("product =" ,result)

find_product(1,2,3,7,8)
```

```python
#odd or even using function

def odd_even(num):
    num = num % 2

    if num > 0:
        print("odd number")
    else:
        print("even number")

num = int(input("Enter the number: "))

odd_even(num)
```

```python
def primeornot(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if primeornot(num):
    print("prime number")
else:
    print("not a prime number")
```

```python
#sum of odd numbers from 1 to 50

sum = 0
for i in range(1, 51):
    if i % 2 == 1:
        sum += i
print("Sum :", sum)
```

```python
num = int(input("Enter a number: "))
result = 1
for i in range(1, num+1):
    result *= i
print("The factorial of", num, "is:", result)
```

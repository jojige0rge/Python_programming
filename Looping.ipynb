1. for loop

```python
fruits = ["apple" ,"banana" ,"cherry"]
for x in fruits:
  print(x)
```

```python
for x in "apple":
  print(x)
```

2. while loop

```python
i = 1
while i <= 6:
  print(i)
  i += 1
```

control statements

```python
#contiue
fruits = ["apple" ,"banana" ,"cherry"]
for x in fruits:
  if x == "apple":
    continue
  print(x)
```

```python
#range
for x in range(2,7):
  print(x)
```

```python
first = 2
last = 500
jump = 5
for x in range(first, last, jump):
    print(x)
```

```python
for x in range(1, 101):
    if x == 51:
        break
    print(x)
```

```python
# prime or not
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print("a prime number")
else:
    print("not a prime number")
```

2. if else

```python
#syntax
if<condition>
  statement
else
  statement
```

```python
num =int(input('enter a number'))
if num > 0:
  print('number is positive')
else:
  print('number is negative')
```

3. elif

```python
#syntax
if<condition>
  statement
elif<condition>
  statement
else
  statement
```

```python
num =int(input('enter a number'))
if num == 0:
  print('zero')
elif num > 0:
  print('number is positive')
else:
  print('number is negative')
```

```python
#odd or even
num = int(input('enter a number:'))
if num % 2 == 0:
  print('even number')
else:
  print('odd number')
```
